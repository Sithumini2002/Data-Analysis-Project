```python
What is the maximum price and which device has this price?
```

```python
row = df.loc[df["Price_USD"].idxmax()]
print("Max Price:", row["Price_USD"])
print("Device Name:", row["Device_Name"])
```

```python
What is the minimum price and which device has this price?
```

```python
row = df.loc[df["Price_USD"].idxmin()]
print("Min Price:", row["Price_USD"])
print("Device Name:", row["Device_Name"])
```

```python
Which brand has the maximum battery life (in hours)?
```

```python
row = df.loc[df["Battery_Life_Hours"].idxmax()]
print("Maximum Battery Life Hours Have Brand Name is:", row["Brand"])
```

```python
Which brand has the minimum battery life (in hours)?
```

```python
row = df.loc[df["Battery_Life_Hours"].idxmin()]
print("Minimum Battery Life Hours Have Brand Name is:", row["Brand"])
```

```python
plt.bar(df["Category"],df["Price_USD"],color="blue")
plt.xlabel("Category Name")
plt.ylabel("Price")
plt.title("Price of Categories")
plt.show()
```

```python
Counts=df["Brand"].value_counts()
Counts
plt.pie(Counts,labels=Counts.index,autopct="%1.1f")
plt.title("Percentage of Brands")
plt.show()
```
